```python
import numpy as np
import os
```

```python
n_data = 500
nt = 250
r = 2
tau = 1
std_err = 1
```

```python
scenarios = ['scenario1','scenario2','scenario3']
```

```python
for scenario in scenarios:
    file_path = './save/simu_data/'+scenario
    
    print('------------ Generate Data:',scenario,'------------')
    
    if os.path.exists(file_path) is False:
        os.makedirs(file_path) 

    if scenario == 'scenario1':
        mu = [0.2,0.2,0.5,0.5]
        cov_mat = np.eye(4)
    elif scenario == 'scenario2':
        mu = np.array([0.2,0.2,0.5,0.5])*np.sqrt(3/2)
        cov_mat = 2*np.eye(4)
    elif scenario == 'scenario3':
        mu = np.array([0.2,0.2,0.5,0.5])*np.sqrt(3/2)
        cov_mat = 1.5*np.eye(4) + np.ones((4,4))*0.5

    params = {'nt':nt,
              'r':r,
              'tau':tau,
              'std_err':std_err,
              'mu':mu,
              'cov_mat':cov_mat}

    for i in range(n_data):
        np.random.seed(i)
        print('data_set:',i)
        x, z = x_gen(nt=nt,r=r,mu=mu,cov_mat=cov_mat)
        y1,y2,y3 = y_gen(x,z,tau,std_err)
        datset = {'x':x,
                  'z':z,
                  'y1':y1,
                  'y2':y2,
                  'y3':y3}
        file_save_path = file_path+'/d'+str(i)+'.npy'
        np.save(file_save_path,datset)
        
        np.save(file_path+'/params.npy',params)
```

```python
np.load(file_path+'/params.npy',allow_pickle=True)
```
