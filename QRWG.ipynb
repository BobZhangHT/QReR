```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from datagen import *
from network import *
```

# Data Generation

```python
nt = 250
r = 2
tau = 1
```

```python
np.random.seed(2020)
x, z = x_gen(nt,r)
y1, y2, y3 = y_gen(x,z,tau)
```

# QRWG Training

QRWG: Quasi-Randomized Weights Generation

```python
batch_size = 256
lr = 1e-3
pa = np.inf
num_nodes = 64
num_iters = 1500
patience = 10
save_folder = './save/'
random_state = 2020
```

```python
estimator = QRWG(lr=lr,
                 batch_size=batch_size,
                 patience=patience,
                 num_iters=num_iters,
                 pa=pa,
                 num_nodes=num_nodes,
                 verbose=True,
                 random_state=random_state)
```

```python
estimator.fit(x,z)
```

# Check Balance

```python
wts_mat_net = estimator.predict().numpy()
```

```python
# consider rerandomization
z_mat_rer = np.array([ReR(pa,torch.Tensor(x))[0].numpy() for i in range(1000)])
```

```python
mdiff_mat_net = np.array([cov_mdiff(x,z,wts_mat_net[i]) for i in range(wts_mat_net.shape[0])])
mdiff_mat_rer = np.array([cov_mdiff(x,z_mat_rer[i]) for i in range(z_mat_rer.shape[0])])
```

```python
import pandas as pd
```

```python
df_mdiff_net = pd.DataFrame(mdiff_mat_net,columns=['X'+str(i+1) for i in range(mdiff_mat_net.shape[1])])
df_mdiff_net['Method'] = 'QRWG'
df_mdiff_rer = pd.DataFrame(mdiff_mat_rer,columns=['X'+str(i+1) for i in range(mdiff_mat_rer.shape[1])])
df_mdiff_rer['Method'] = 'CR'
```

```python
df_mdiff = pd.concat([df_mdiff_net,df_mdiff_rer],axis=0)
df_mdiff = pd.melt(df_mdiff,id_vars=['Method'],var_name=['Variable'],value_name='Value')
```

```python
plt.figure(figsize=(10,6))
sb.boxplot(x='Variable',y='Value',
           hue="Method",data=df_mdiff)
plt.ylabel('$\\bar{x}_T-\\bar{x}_C$',fontsize=12)
plt.xlabel('Covariates',fontsize=12)
```

# Inference

```python
from datagen import tau_diff
tau_vec_net = np.array([tau_diff(y1,z,wts_mat_net[i]) for i in range(wts_mat_net.shape[0])])
```

```python
# mean
tau_vec_net.mean()
```

```python
# confidence interval
np.quantile(tau_vec_net,(0.025,0.975))
```

```python
tau_vec_rer = []
for i in range(z_mat_rer.shape[0]):
    y1_rer,_,_ = y_gen(x,z_mat_rer[i],tau=tau)
    tau_vec_rer.append(tau_diff(y1_rer,z_mat_rer[i]))
tau_vec_rer = np.array(tau_vec_rer)
```

```python
plt.figure()
sb.distplot(tau_vec_rer,label='ydiff_rer')
sb.distplot(tau_vec_net,label='ydiff_qrwg')
plt.legend()
```
