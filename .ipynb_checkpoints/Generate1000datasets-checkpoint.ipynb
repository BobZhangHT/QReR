```python
from datagen import *
```

```python
import numpy as np
import os
```

```python
n_data = 1000
```

```python
file_path = './save/simu_data'
```

```python
if os.path.exists(file_path) is False:
    os.makedirs(file_path) 
```

```python
nt = 250
r = 2
tau = 1
std_err = 1
```

```python
params = {'nt':nt,
          'r':r,
          'tau':tau,
          'std_err':std_err}
```

```python
for i in range(n_data):
    np.random.seed(i)
    print('data_set:',i)
    x, z = x_gen(nt,r)
    y1,y2,y3 = y_gen(x,z,tau,std_err)
    datset = {'x':x,
          'z':z,
          'y1':y1,
          'y2':y2,
          'y3':y3}
    file_save_path = file_path+'/d'+str(i)+'.npy'
    np.save(file_save_path,datset)
```

```python
np.save('./save/simu_data/params.npy',params)
```

```python
np.load('./save/simu_data/params.npy',allow_pickle=True)
```
