```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

from datagen import x_gen,y_gen
from network import *
```

# Data Generation

```python
nt = 250
r = 2
tau = 1
```

```python
np.random.seed(2020)
x, z = x_gen(nt,r)
y1, y2, y3 = y_gen(x,z,tau)
```

# QRWG Training

QRWG: Quasi-Randomized Weights Generation

```python
batch_size = 256
lr = 1e-3
pa = np.inf
num_nodes = 64
num_iters = 1500
patience = 10
save_folder = './save/'
random_state = 2020
```

```python
estimator = QRWG(lr=lr,
                 batch_size=batch_size,
                 patience=patience,
                 num_iters=num_iters,
                 pa=pa,
                 num_nodes=num_nodes,
                 verbose=True,
                 random_state=random_state)
```

```python
estimator.fit(x,z)
```

# Inference

```python
wts_mat_net = estimator.predict()
```

```python
tau_vec_net = np.array([tau_diff(y1,z,wts_mat_net[i]) for i in range(wts_mat_net.shape[0])])
```

```python
# mean
tau_vec_net.mean()
```

```python
# confidence interval
np.quantile(tau_vec_net,(0.025,0.975))
```

```python
# consider rerandomization
z_mat_rer = np.array([ReR(pa,x)[0].numpy() for i in range(1000)])
```

```python
tau_vec_rer = []
for i in range(z_mat_rer.shape[0]):
    y1_rer,_ = y_gen(X,z_mat_rer[i],tau=tau)
    tau_vec_rer.append(tau_diff(y1_rer,z_mat_rer[i]))
tau_vec_rer = np.array(tau_vec_rer)
```

```python
plt.figure()
sb.distplot(tau_vec_rer,label='ydiff_rer')
sb.distplot(tau_vec_net,label='ydiff_qrwg')
plt.legend()
```
